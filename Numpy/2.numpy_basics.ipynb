# NumPy Basics
A hands-on reference notebook covering the most common NumPy operations.

**Topics covered:**
1. Array creation
2. Array attributes
3. Indexing & slicing
4. Reshaping
5. Math operations
6. Aggregations
7. Broadcasting
8. Boolean masking
9. Stacking & splitting
10. Useful utility functions

```python
import numpy as np
print('NumPy version:', np.__version__)
```

## 1. Array Creation

```python
# From a Python list
a = np.array([1, 2, 3, 4, 5])
print('1D array:', a)

# 2D array (matrix)
b = np.array([[1, 2, 3],
              [4, 5, 6]])
print('2D array:\n', b)

# Special arrays
print('zeros (3x3):\n', np.zeros((3, 3)))
print('ones (2x4):\n',  np.ones((2, 4)))
print('identity (3x3):\n', np.eye(3))
print('full (2x3, value=7):\n', np.full((2, 3), 7))

# Range-based
print('arange(0,10,2):', np.arange(0, 10, 2))
print('linspace(0,1,5):', np.linspace(0, 1, 5))

# Random
rng = np.random.default_rng(seed=42)   # reproducible
print('random floats:', rng.random((2, 3)))
print('random ints:', rng.integers(0, 10, size=(2, 3)))
```

## 2. Array Attributes

```python
arr = np.array([[1, 2, 3], [4, 5, 6]], dtype=np.float32)

print('shape  :', arr.shape)      # (rows, cols)
print('ndim   :', arr.ndim)       # number of dimensions
print('size   :', arr.size)       # total elements
print('dtype  :', arr.dtype)      # data type
print('itemsize:', arr.itemsize)  # bytes per element
print('nbytes :', arr.nbytes)     # total bytes
```

## 3. Indexing & Slicing

```python
m = np.arange(12).reshape(3, 4)
print('matrix:\n', m)

# Single element
print('m[1, 2]  =', m[1, 2])          # row 1, col 2

# Slices
print('m[0, :]  =', m[0, :])           # first row
print('m[:, 2]  =', m[:, 2])           # third column
print('m[1:, 1:3] =\n', m[1:, 1:3])   # sub-matrix

# Fancy (integer array) indexing
rows = np.array([0, 2])
cols = np.array([1, 3])
print('fancy index:', m[rows, cols])    # m[0,1] and m[2,3]

# Negative indexing
print('last row   :', m[-1])
print('last col   :', m[:, -1])
```

## 4. Reshaping & Transposing

```python
a = np.arange(12)
print('flat:', a)

b = a.reshape(3, 4)
print('reshape (3,4):\n', b)

# -1 infers the missing dimension
c = a.reshape(2, -1)   # 2 rows, numpy works out cols
print('reshape (2,-1):\n', c)

# Transpose
print('transpose:\n', b.T)

# Flatten vs ravel
print('flatten (copy):',  b.flatten())   # always a copy
print('ravel   (view):',  b.ravel())     # view when possible

# Add/remove axes
v = np.array([1, 2, 3])
print('newaxis -> col vector:\n', v[:, np.newaxis])
print('squeeze removes size-1 axes:', np.squeeze(v[:, np.newaxis]).shape)
```

## 5. Math Operations

```python
x = np.array([1, 2, 3, 4], dtype=float)
y = np.array([10, 20, 30, 40], dtype=float)

# Element-wise arithmetic (vectorised — no Python loops needed)
print('x + y  :', x + y)
print('x * y  :', x * y)
print('x ** 2 :', x ** 2)
print('y / x  :', y / x)

# Universal functions (ufuncs)
print('sqrt   :', np.sqrt(x))
print('exp    :', np.exp(x))
print('log    :', np.log(y))
print('sin    :', np.sin(np.linspace(0, np.pi, 5)))
print('abs    :', np.abs(np.array([-3, -1, 0, 2, 4])))

# Matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print('A @ B (matmul):\n', A @ B)
print('dot product of vectors:', np.dot(x, y))
```

## 6. Aggregation Functions

```python
m = np.array([[4, 7, 2],
              [1, 9, 5],
              [8, 3, 6]])
print('matrix:\n', m)

# Global aggregations
print('sum  :', m.sum())
print('min  :', m.min(),  '  max:', m.max())
print('mean :', m.mean(), '  std:', m.std().round(3))
print('median:', np.median(m))

# axis=0 → collapse rows (operate down columns)
# axis=1 → collapse cols (operate across rows)
print('col sums (axis=0):', m.sum(axis=0))
print('row sums (axis=1):', m.sum(axis=1))
print('col min  (axis=0):', m.min(axis=0))

# Index of min/max
print('argmax (flat):', m.argmax())
print('argmax per row:', m.argmax(axis=1))

# Cumulative
print('cumsum (flat):', m.cumsum())
print('cumprod row0 :', m[0].cumprod())
```

## 7. Broadcasting

```python
# Broadcasting lets arrays with different shapes work together
# Rule: dimensions are compatible if they are equal OR one of them is 1.

A = np.array([[1, 2, 3],   # shape (3, 3)
              [4, 5, 6],
              [7, 8, 9]])

v = np.array([10, 20, 30]) # shape (3,)  ->  broadcast to (3, 3)

print('A + v (add row vector to every row):\n', A + v)

col = np.array([[10],      # shape (3, 1)  ->  broadcast to (3, 3)
                [20],
                [30]])
print('A + col (add col vector to every col):\n', A + col)

# Practical: zero-centre each column
centred = A - A.mean(axis=0)
print('column-centred A:\n', centred)
```

## 8. Boolean Masking & Filtering

```python
data = np.array([15, 3, 22, 8, 41, 7, 19, 30])

# Boolean mask
mask = data > 10
print('mask          :', mask)
print('filtered data :', data[mask])

# Inline
print('values < 20   :', data[data < 20])

# Compound conditions (use & | ~, NOT 'and' 'or')
print('10 < val < 25 :', data[(data > 10) & (data < 25)])

# np.where(condition, val_if_true, val_if_false)
clipped = np.where(data > 20, 20, data)
print('clipped at 20 :', clipped)

# Count / any / all
print('count > 10    :', np.sum(data > 10))
print('any > 40      :', np.any(data > 40))
print('all > 0       :', np.all(data > 0))
```

## 9. Stacking & Splitting

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# Stack
print('vstack (row-wise):\n',  np.vstack([a, b]))   # (4,2)
print('hstack (col-wise):\n',  np.hstack([a, b]))   # (2,4)
print('dstack (depth)   :\n',  np.dstack([a, b]))   # (2,2,2)
print('concatenate axis=0:\n', np.concatenate([a, b], axis=0))
print('concatenate axis=1:\n', np.concatenate([a, b], axis=1))

# Split
x = np.arange(12).reshape(4, 3)
parts = np.vsplit(x, 2)      # split into 2 equal halves along rows
print('vsplit into 2:');
[print(' ', p) for p in parts]

cols = np.hsplit(x, 3)       # split into 3 equal halves along cols
print('hsplit into 3:');
[print(' ', c.T) for c in cols]
```

## 10. Useful Utilities

```python
# Sorting
arr = np.array([3, 1, 4, 1, 5, 9, 2, 6])
print('sorted           :', np.sort(arr))       # returns sorted copy
print('argsort (indices):', np.argsort(arr))    # indices that would sort

# Unique values
print('unique           :', np.unique(arr))
vals, counts = np.unique(arr, return_counts=True)
print('value counts     :', dict(zip(vals, counts)))

# Clipping
print('clipped [2,6]    :', np.clip(arr, 2, 6))

# Type casting
floats = arr.astype(np.float64)
print('as float64       :', floats)

# Copying vs views
view = arr[2:5]      # slice is a VIEW — modifying it changes arr!
copy = arr[2:5].copy()
view[0] = 99
print('after view[0]=99, arr:', arr)   # arr changed
copy[0] = 99
print('copy is independent :', copy)   # arr unchanged

# Rounding & precision
x = np.array([1.2345, 2.6789, 3.5])
print('round 2dp        :', np.round(x, 2))
print('floor / ceil     :', np.floor(x), '/', np.ceil(x))
```
