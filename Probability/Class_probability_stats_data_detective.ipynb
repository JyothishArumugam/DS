
## 🪙 Case #1: The Coin Flip Club

Your first assignment is almost embarrassingly simple: a group at the Academy claims they have a coin that is **perfectly fair**. Your job is to verify it — not by *arguing*, but by *flipping it, a lot, and watching what happens*.

### The core idea

For any experiment with equally likely outcomes:

$$P(\text{Event}) = \frac{\text{Number of favorable outcomes}}{\text{Total number of possible outcomes}}$$

For a fair coin, $P(\text{Heads}) = \frac{1}{2} = 0.5$. That's the **theoretical probability** — a promise about what *should* happen in the long run.

But real life is a single run of tosses. What you actually observe is the **empirical (experimental) probability**:

$$\hat{P}(\text{Heads}) = \frac{\text{Number of Heads observed}}{\text{Total tosses}}$$

The big question detectives care about: *as you flip more and more times, does the empirical probability converge to the theoretical one?* This is called the **Law of Large Numbers** — and it's the bedrock of everything that follows in this notebook.

Use the slider below to flip the coin anywhere from 1 to 5,000 times and watch what happens.


```python

def coin_flip_experiment(n_tosses=50, true_p_heads=0.5):
    flips = np.random.binomial(1, true_p_heads, size=n_tosses)  # 1 = Heads, 0 = Tails
    running_prob = np.cumsum(flips) / np.arange(1, n_tosses + 1)

    fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

    # Left: running empirical probability vs true probability
    axes[0].plot(running_prob, color='#1f77b4', lw=1.5, label='Empirical P(Heads)')
    axes[0].axhline(true_p_heads, color='red', ls='--', label=f'Theoretical P(Heads) = {true_p_heads}')
    axes[0].set_xlabel('Number of tosses')
    axes[0].set_ylabel('P(Heads) so far')
    axes[0].set_title('Law of Large Numbers in action')
    axes[0].set_ylim(0, 1)
    axes[0].legend()

    # Right: bar chart of final counts
    heads = flips.sum()
    tails = n_tosses - heads
    axes[1].bar(['Heads', 'Tails'], [heads, tails], color=['#2ca02c', '#d62728'])
    axes[1].set_title(f'After {n_tosses} tosses: {heads} Heads, {tails} Tails')
    axes[1].set_ylabel('Count')

    plt.tight_layout()
    plt.show()

    print(f"Empirical P(Heads) = {heads/n_tosses:.4f}   |   Theoretical P(Heads) = {true_p_heads}")

interact(
    coin_flip_experiment,
    n_tosses=widgets.IntSlider(min=1, max=5000, step=10, value=50, description='Tosses:'),
    true_p_heads=widgets.FloatSlider(min=0.05, max=0.95, step=0.05, value=0.5, description='True P(Heads):')
);
```


**Detective's notebook 📝**

- Drag `n_tosses` up to 5000 and watch the blue line settle down near the red dashed line. That's the Law of Large Numbers — small samples are noisy, big samples are trustworthy.
- Now try `true_p_heads = 0.7` — this is a **biased coin**. Notice the empirical line still finds its way to the true value, just not to 0.5.

## 🏥 Case #2: The Clinic Files

The Academy runs a small clinic that screens trainees for a (fictional) condition called **Flu-X**. You're handed a test report and asked a deceptively simple question:

> *"If a trainee tests positive, what's the probability they actually have Flu-X?"*

This is fundamentally different from Case #1. Now we care about **two events happening together**, and specifically how knowing one changes what we believe about the other.

### The core idea

**Conditional probability** — the probability of event $A$, *given that* event $B$ has already happened:

$$P(A \mid B) = \frac{P(A \cap B)}{P(B)}$$

In plain English: *"Out of all the times B happened, in what fraction of them did A also happen?"*

Rather than jump straight to formulas, let's **simulate an entire population** of trainees and literally count them.


```python

def clinic_simulation(population_size=10000, prevalence=0.05, sensitivity=0.9, specificity=0.9):
    # prevalence  = P(Disease+)      -- how common Flu-X actually is
    # sensitivity = P(Test+ | Disease+)  -- test correctly catches sick people
    # specificity = P(Test- | Disease-)  -- test correctly clears healthy people

    has_disease = np.random.rand(population_size) < prevalence

    test_positive = np.empty(population_size, dtype=bool)
    # sick people: positive test with probability = sensitivity
    sick_idx = has_disease
    test_positive[sick_idx] = np.random.rand(sick_idx.sum()) < sensitivity
    # healthy people: positive test (false alarm) with probability = 1 - specificity
    healthy_idx = ~has_disease
    test_positive[healthy_idx] = np.random.rand(healthy_idx.sum()) < (1 - specificity)

    # Build the 2x2 contingency table
    tp = np.sum(has_disease & test_positive)
    fn = np.sum(has_disease & ~test_positive)
    fp = np.sum(~has_disease & test_positive)
    tn = np.sum(~has_disease & ~test_positive)

    table = pd.DataFrame(
        [[tp, fn, tp + fn], [fp, tn, fp + tn], [tp + fp, fn + tn, population_size]],
        index=['Disease +', 'Disease -', 'Total'],
        columns=['Test +', 'Test -', 'Total']
    )
    display(table)

    p_disease_given_positive = tp / (tp + fp) if (tp + fp) > 0 else float('nan')
    p_positive_given_disease = tp / (tp + fn) if (tp + fn) > 0 else float('nan')

    print(f"\nP(Test + | Disease +)  [this is the SENSITIVITY you set] = {p_positive_given_disease:.3f}")
    print(f"P(Disease + | Test +)  [THE ANSWER YOU ACTUALLY WANT]     = {p_disease_given_positive:.3f}")
    print(f"\nOut of {tp+fp} trainees who tested positive, only {tp} of them are actually sick.")

interact(
    clinic_simulation,
    population_size=widgets.fixed(10000),
    prevalence=widgets.FloatSlider(min=0.01, max=0.5, step=0.01, value=0.05, description='Prevalence:'),
    sensitivity=widgets.FloatSlider(min=0.5, max=0.99, step=0.01, value=0.9, description='Sensitivity:'),
    specificity=widgets.FloatSlider(min=0.5, max=0.99, step=0.01, value=0.9, description='Specificity:')
);
```


**Detective's notebook 📝**

Notice something strange: even with a test that's "90% accurate" in both directions, $P(\text{Disease} \mid \text{Test+})$ can be surprisingly *low* when the disease is rare (`prevalence` is small). Testing positive is not the same as being sick — those are two different conditional probabilities, and mixing them up is one of the most common reasoning mistakes people make (in medicine, in court cases, and in everyday life).

Try dragging `prevalence` down to 0.01 and watch `P(Disease + | Test +)` fall — even though sensitivity and specificity haven't changed at all.

This exact confusion is what our next case is built around.



## 🔍 Case #3: The False Alarm

A trainee bursts into your office, panicked: *"I tested positive for Flu-X! Am I going to be okay?!"*

You now know enough to ask the right question back: *"How rare is Flu-X in the first place?"* That single piece of information — your **prior belief** before seeing any evidence — turns out to matter enormously.

### The core idea

We already simulated $P(\text{Disease} \mid \text{Test+})$ by brute-force counting people. **Bayes' Theorem** gives us a formula to compute it directly, without simulating an entire population:

$$P(A \mid B) = \frac{P(B \mid A)\, P(A)}{P(B)}$$

Mapped onto our clinic story:

$$P(\text{Disease+} \mid \text{Test+}) = \frac{P(\text{Test+} \mid \text{Disease+}) \cdot P(\text{Disease+})}{P(\text{Test+})}$$

where the denominator expands via the **law of total probability**:

$$P(\text{Test+}) = \underbrace{P(\text{Test+}\mid\text{Disease+})P(\text{Disease+})}_{\text{true positives}} + \underbrace{P(\text{Test+}\mid\text{Disease-})P(\text{Disease-})}_{\text{false positives}}$$

Each piece has a name detectives use constantly:
- $P(\text{Disease+})$ — the **prior**: what you believed before any evidence.
- $P(\text{Test+} \mid \text{Disease+})$ — the **likelihood**: how well the evidence fits the hypothesis.
- $P(\text{Disease+} \mid \text{Test+})$ — the **posterior**: your *updated* belief after seeing the evidence.

**Bayes' Theorem is just a precise recipe for updating beliefs with evidence** — which is the whole job description of a detective.


```python

def bayes_theorem_explorer(prior_prevalence=0.05, sensitivity=0.9, specificity=0.9):
    p_disease = prior_prevalence
    p_no_disease = 1 - prior_prevalence

    p_pos_given_disease = sensitivity
    p_pos_given_no_disease = 1 - specificity

    # Law of total probability
    p_positive = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)

    # Bayes' theorem
    posterior = (p_pos_given_disease * p_disease) / p_positive

    fig, ax = plt.subplots(figsize=(7, 4.5))
    bars = ax.bar(['Prior belief\nP(Disease+)', 'Posterior belief\nP(Disease+ | Test+)'],
                   [p_disease, posterior], color=['#7f7f7f', '#d62728'])
    ax.set_ylim(0, 1)
    ax.set_ylabel('Probability')
    ax.set_title('How one positive test result updates your belief')
    for bar, val in zip(bars, [p_disease, posterior]):
        ax.text(bar.get_x() + bar.get_width()/2, val + 0.02, f'{val:.3f}', ha='center', fontweight='bold')
    plt.tight_layout()
    plt.show()

    print(f"Before the test: you believed there was a {p_disease*100:.1f}% chance of Flu-X.")
    print(f"After one positive test: that belief updates to {posterior*100:.1f}%.")
    print(f"(P(Test+) overall, healthy or sick, = {p_positive:.3f})")

interact(
    bayes_theorem_explorer,
    prior_prevalence=widgets.FloatSlider(min=0.001, max=0.5, step=0.001, value=0.05, readout_format='.3f', description='Prior:'),
    sensitivity=widgets.FloatSlider(min=0.5, max=0.999, step=0.001, value=0.9, description='Sensitivity:'),
    specificity=widgets.FloatSlider(min=0.5, max=0.999, step=0.001, value=0.9, description='Specificity:')
);
```


**Detective's notebook 📝**

- Compare these numbers to Case #2 — they should match your simulation almost exactly (simulation is noisy, Bayes' Theorem is exact).
- Push `prior_prevalence` down toward 0.001 (a genuinely rare disease). Even a fairly accurate test barely moves the needle — this is the famous **base rate fallacy**, and it's the reason doctors often order a *second, different* test before treating a positive result as certain.

## 📊 Case #4: The Trainee Scores

The Academy just ran its entrance exam for 200 new recruits. Before you can say anything meaningful about *how the class did*, you need to **summarize** the data — this is the job of **descriptive statistics**.

### The core ideas

**Measures of central tendency** (where's the "middle" of the data?):
- **Mean** $\bar{x} = \frac{1}{n}\sum x_i$ — the arithmetic average. Sensitive to extreme values.
- **Median** — the middle value when sorted. Robust to extreme values.
- **Mode** — the most frequently occurring value.

**Measures of spread** (how "spread out" is the data?):
- **Variance** $\sigma^2 = \frac{1}{n}\sum (x_i - \bar{x})^2$ — average squared distance from the mean.
- **Standard deviation** $\sigma = \sqrt{\sigma^2}$ — same units as the data, easier to interpret.

Let's generate a realistic batch of exam scores and see these numbers come alive — including what happens when a single unusual score (an **outlier**) sneaks into the data.


```python

np.random.seed(1)
base_scores = np.clip(np.random.normal(loc=68, scale=10, size=200), 0, 100)

def descriptive_stats_explorer(outlier_value=68, add_outlier=False):
    scores = base_scores.copy()
    if add_outlier:
        scores = np.append(scores, outlier_value)

    mean_val = np.mean(scores)
    median_val = np.median(scores)
    mode_val = stats.mode(np.round(scores), keepdims=True).mode[0]
    std_val = np.std(scores)
    var_val = np.var(scores)

    fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

    axes[0].hist(scores, bins=20, color='#1f77b4', edgecolor='white', alpha=0.85)
    axes[0].axvline(mean_val, color='red', ls='--', label=f'Mean = {mean_val:.1f}')
    axes[0].axvline(median_val, color='green', ls='--', label=f'Median = {median_val:.1f}')
    axes[0].set_title('Distribution of Exam Scores')
    axes[0].set_xlabel('Score')
    axes[0].legend()

    axes[1].boxplot(scores, vert=True)
    axes[1].set_title('Boxplot (spot the outliers)')
    axes[1].set_ylabel('Score')

    plt.tight_layout()
    plt.show()

    print(f"n = {len(scores)}")
    print(f"Mean               = {mean_val:.2f}")
    print(f"Median             = {median_val:.2f}")
    print(f"Mode (rounded)     = {mode_val:.2f}")
    print(f"Standard deviation = {std_val:.2f}")
    print(f"Variance           = {var_val:.2f}")

interact(
    descriptive_stats_explorer,
    add_outlier=widgets.Checkbox(value=False, description='Add one weird score'),
    outlier_value=widgets.IntSlider(min=0, max=100, step=1, value=100, description='Outlier score:')
);
```


**Detective's notebook 📝**

- With `add_outlier` off, mean and median should sit close together — the data is roughly symmetric.
- Now check `add_outlier`, and drag the outlier score to 0 or 100. Watch the **mean** jump around while the **median barely moves**. This is exactly why news reports on "average income" can be misleading — a few billionaires drag the mean way above what a typical person earns, while the median stays representative.
- **Exercise:** Set the outlier to exactly the mean value (so it *shouldn't* change anything) — confirm the mean and median stay put.

## 🔔 Case #5: The Bell Curve Case

Heights, exam scores, measurement errors, blood pressure readings — an enormous number of real-world quantities cluster around an average in a very particular bell-shaped pattern. That pattern is the **Gaussian (Normal) distribution**, and it's arguably the single most important distribution in all of statistics.

### The core idea

The normal distribution is fully described by just two numbers — its mean $\mu$ (center) and standard deviation $\sigma$ (spread):

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} \, e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

You don't need to memorize this formula. What you *do* need is intuition for the **68-95-99.7 rule**:
- ~68% of values fall within $\mu \pm 1\sigma$
- ~95% of values fall within $\mu \pm 2\sigma$
- ~99.7% of values fall within $\mu \pm 3\sigma$

Play with $\mu$, $\sigma$, and a target range below to see exactly how much probability lives under different parts of the curve.


```python

def gaussian_explorer(mu=0.0, sigma=1.0, lower=-1.0, upper=1.0):
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
    y = stats.norm.pdf(x, mu, sigma)

    fig, ax = plt.subplots(figsize=(9, 4.5))
    ax.plot(x, y, color='#1f77b4', lw=2)

    fill_x = np.linspace(max(lower, mu - 4*sigma), min(upper, mu + 4*sigma), 300)
    ax.fill_between(fill_x, stats.norm.pdf(fill_x, mu, sigma), color='#1f77b4', alpha=0.35)

    ax.axvline(mu, color='black', ls=':', alpha=0.6, label=f'mean = {mu}')
    ax.set_title(f'Normal Distribution  (μ={mu}, σ={sigma})')
    ax.set_xlabel('x')
    ax.set_ylabel('Probability density')
    ax.legend()
    plt.tight_layout()
    plt.show()

    prob = stats.norm.cdf(upper, mu, sigma) - stats.norm.cdf(lower, mu, sigma)
    print(f"P({lower} < X < {upper}) = {prob:.4f}   (that's {prob*100:.2f}% of the area under the curve)")

interact(
    gaussian_explorer,
    mu=widgets.FloatSlider(min=-10, max=10, step=0.5, value=0, description='Mean (μ):'),
    sigma=widgets.FloatSlider(min=0.5, max=5, step=0.5, value=1, description='Std dev (σ):'),
    lower=widgets.FloatSlider(min=-20, max=20, step=0.5, value=-1, description='Lower bound:'),
    upper=widgets.FloatSlider(min=-20, max=20, step=0.5, value=1, description='Upper bound:')
);
```


**Detective's notebook 📝**

- Set `lower = μ - σ` and `upper = μ + σ` — you should get roughly **0.68**, confirming the 68-95-99.7 rule.
- Widen it to $\pm2\sigma$ — you should land near **0.95**. This exact number is why "95% confidence" shows up everywhere in statistics — spoiler for Case #7.
- **Exercise:** If exam scores are normal with $\mu=70, \sigma=10$, what fraction of trainees scored above 90? Set the sliders to check your intuition.

## 🌀 Case #6: The Rookie Paradox

Here's the puzzle: the Academy's obstacle course completion **times** are wildly skewed — most rookies finish reasonably fast, but a stubborn few take forever, dragging out a long tail. This data is nowhere close to a bell curve.

And yet — if you take *groups* of rookies and look at each group's **average** time, something remarkable happens.

### The core idea — Central Limit Theorem (CLT)

> No matter what the shape of the original population is, if you repeatedly draw random samples of size $n$ and compute their **means**, the distribution of those sample means approaches a **Normal distribution** as $n$ grows — with:
>
> $$\text{mean of sample means} = \mu \qquad \text{std dev of sample means} = \frac{\sigma}{\sqrt{n}}$$

This is arguably the most important theorem in all of statistics, because it means: even when we know nothing about the underlying population, we can still use the well-understood normal distribution to reason about **averages**.

Pick a wildly non-normal population below, choose a sample size, and watch the sampling distribution of the mean reshape itself into a bell curve as $n$ grows.


```python

def clt_explorer(population='Exponential (skewed)', sample_size=5, num_samples=2000):
    rng = np.random.default_rng(7)

    if population == 'Exponential (skewed)':
        pop = rng.exponential(scale=2.0, size=100000)
    elif population == 'Uniform (flat)':
        pop = rng.uniform(0, 10, size=100000)
    else:  # Bimodal
        pop = np.concatenate([rng.normal(2, 0.5, 50000), rng.normal(8, 0.5, 50000)])

    sample_means = np.array([
        rng.choice(pop, size=sample_size, replace=True).mean()
        for _ in range(num_samples)
    ])

    fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

    axes[0].hist(pop, bins=60, color='#7f7f7f', alpha=0.8)
    axes[0].set_title(f'Original Population\n({population}) — NOT normal')
    axes[0].set_xlabel('Value')

    axes[1].hist(sample_means, bins=40, color='#ff7f0e', alpha=0.85, density=True)
    x = np.linspace(sample_means.min(), sample_means.max(), 200)
    axes[1].plot(x, stats.norm.pdf(x, sample_means.mean(), sample_means.std()), color='blue', lw=2, label='Normal fit')
    axes[1].set_title(f'Distribution of Sample Means\n(n={sample_size} per sample, {num_samples} samples)')
    axes[1].set_xlabel('Sample mean')
    axes[1].legend()

    plt.tight_layout()
    plt.show()

    print(f"Population mean (μ)        = {pop.mean():.3f}")
    print(f"Mean of sample means       = {sample_means.mean():.3f}  (should be close to μ)")
    print(f"Population std / sqrt(n)   = {pop.std()/np.sqrt(sample_size):.3f}")
    print(f"Std dev of sample means    = {sample_means.std():.3f}  (should be close to the line above)")

interact(
    clt_explorer,
    population=widgets.Dropdown(options=['Exponential (skewed)', 'Uniform (flat)', 'Bimodal'], description='Population:'),
    sample_size=widgets.IntSlider(min=1, max=200, step=1, value=5, description='Sample size (n):'),
    num_samples=widgets.IntSlider(min=100, max=5000, step=100, value=2000, description='# of samples:')
);
```


**Detective's notebook 📝**

- Start with `sample_size = 1`. The right-hand histogram should look *identical* to the population — makes sense, a "sample" of 1 is just a single raw value.
- Now push `sample_size` up to 30, then 100. Watch the orange histogram morph into a clean bell curve, **regardless of which population you picked**.
- **Exercise:** Compare "std dev of sample means" printed at the bottom against $\sigma/\sqrt{n}$ — this relationship is called the **Standard Error of the Mean**, and it's the reason larger samples give you more precise estimates.

The CLT is the bridge that lets us go from "raw, messy data" to "confident numerical claims." Let's use it to quantify exactly how confident.



## 🎯 Case #7: How Sure Are We?

You've measured the average reaction time of 40 rookies and gotten a number. But you know from Case #6 that if you resampled a *different* 40 rookies, you'd get a slightly different average. So how do you honestly report your result?

The answer isn't a single number — it's a **range**, paired with a **confidence level**.

### The core idea

A $C\%$ **confidence interval** for the mean is:

$$\bar{x} \pm z^{*} \cdot \frac{s}{\sqrt{n}}$$

where $\bar{x}$ is your sample mean, $s$ is your sample standard deviation, $n$ is your sample size, and $z^{*}$ is a multiplier from the normal distribution (e.g. $z^{*}=1.96$ for 95% confidence).

**The correct interpretation (this trips almost everyone up the first time):**

> A 95% confidence interval does **not** mean "there's a 95% chance the true mean is in this specific interval." It means: *"if we repeated this whole sampling process many times and built an interval every time, about 95% of those intervals would contain the true population mean."*

Let's not just say that — let's **watch it happen** by running the experiment hundreds of times.


```python

TRUE_MEAN = 250.0   # true average reaction time in ms (unknown to the "detective" in real life)
TRUE_STD = 30.0

def confidence_interval_explorer(confidence_level=95, sample_size=40, num_experiments=100):
    rng = np.random.default_rng(0)
    z = stats.norm.ppf(0.5 + confidence_level/200)

    contains_true_mean = 0
    fig, ax = plt.subplots(figsize=(9, 6))

    for i in range(num_experiments):
        sample = rng.normal(TRUE_MEAN, TRUE_STD, size=sample_size)
        x_bar = sample.mean()
        margin = z * sample.std(ddof=1) / np.sqrt(sample_size)
        lo, hi = x_bar - margin, x_bar + margin
        hit = lo <= TRUE_MEAN <= hi
        contains_true_mean += hit
        color = '#2ca02c' if hit else '#d62728'
        ax.plot([lo, hi], [i, i], color=color, alpha=0.6, lw=1.5)

    ax.axvline(TRUE_MEAN, color='black', ls='--', lw=2, label=f'True mean = {TRUE_MEAN}')
    ax.set_xlabel('Reaction time (ms)')
    ax.set_ylabel('Experiment #')
    ax.set_title(f'{num_experiments} independent {confidence_level}% confidence intervals\n'
                 f'(green = contains true mean, red = misses it)')
    ax.legend()
    plt.tight_layout()
    plt.show()

    print(f"Out of {num_experiments} experiments, {contains_true_mean} "
          f"({contains_true_mean/num_experiments*100:.1f}%) of the intervals captured the true mean.")
    print(f"Target was ~{confidence_level}% by design.")

interact(
    confidence_interval_explorer,
    confidence_level=widgets.SelectionSlider(options=[80, 90, 95, 99], value=95, description='Confidence %:'),
    sample_size=widgets.IntSlider(min=5, max=200, step=5, value=40, description='Sample size:'),
    num_experiments=widgets.IntSlider(min=20, max=300, step=10, value=100, description='# experiments:')
);
```


**Detective's notebook 📝**

- Run it at 95% confidence and count the red lines — you should see roughly 5 misses out of 100, matching the theory almost exactly.
- Drop to 80% confidence — more red lines. Higher confidence requires a *wider* net, which is the fundamental precision-vs-confidence trade-off.
- Increase `sample_size` — intervals get narrower for the *same* confidence level. Bigger samples → more precise estimates (straight from the CLT in Case #6).

## 🎯 Case #9... wait, Case #8: The Streak Counter

Back to basics — remember the fair coin from Case #1? Now the question changes shape: *"If I flip a coin 10 times, what's the probability I get **exactly** 7 heads?"*

This is a **Binomial** question: a fixed number of independent trials, each with the same probability of "success."

### The core idea

If $X$ = number of successes in $n$ independent trials, each with success probability $p$:

$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}, \qquad \binom{n}{k} = \frac{n!}{k!(n-k)!}$$

- Mean: $E[X] = np$
- Variance: $\text{Var}(X) = np(1-p)$

This is exactly the same coin-flipping machinery from Case #1 — we're just now asking about the *count* of heads instead of watching the running probability evolve.


```python

def binomial_explorer(n=10, p=0.5, highlight_k=7):
    k_values = np.arange(0, n + 1)
    pmf = stats.binom.pmf(k_values, n, p)

    fig, ax = plt.subplots(figsize=(9, 4.5))
    colors = ['#d62728' if k == highlight_k else '#1f77b4' for k in k_values]
    ax.bar(k_values, pmf, color=colors, edgecolor='white')
    ax.set_xlabel('Number of successes (k)')
    ax.set_ylabel('P(X = k)')
    ax.set_title(f'Binomial Distribution:  n={n} trials, p={p}')
    plt.tight_layout()
    plt.show()

    print(f"P(X = {highlight_k}) = {stats.binom.pmf(highlight_k, n, p):.4f}")
    print(f"P(X <= {highlight_k}) = {stats.binom.cdf(highlight_k, n, p):.4f}")
    print(f"Expected value E[X] = n*p = {n*p:.2f}")
    print(f"Variance = n*p*(1-p) = {n*p*(1-p):.2f}")

interact(
    binomial_explorer,
    n=widgets.IntSlider(min=1, max=100, step=1, value=10, description='Trials (n):'),
    p=widgets.FloatSlider(min=0.01, max=0.99, step=0.01, value=0.5, description='P(success):'),
    highlight_k=widgets.IntSlider(min=0, max=100, step=1, value=7, description='Highlight k:')
);
```


**Detective's notebook 📝**

- Set `n=10, p=0.5` and check `highlight_k=5` — the peak (most likely outcome). Makes sense: with a fair coin, 5 heads out of 10 is the most common result.
- Push `n` up toward 100 while keeping `p=0.5`. Notice the whole distribution starts looking suspiciously like... a bell curve. That's not a coincidence — the Binomial distribution *approaches* the Normal distribution for large $n$ (another flavor of the CLT from Case #6!).

## ⏳ Case #9: The Waiting Game

Final case, rookie. Instead of asking "how many successes in $n$ tries," we flip the question: *"how many tries until the **first** success?"*

Imagine you're testing a lock-picking trainee who succeeds with probability $p$ on each attempt, and you want to know how many attempts it typically takes to get their **first** success.

### The core idea

If $X$ = number of trials needed to get the first success:

$$P(X = k) = (1-p)^{k-1} \, p, \qquad k = 1, 2, 3, \dots$$

- Mean: $E[X] = \frac{1}{p}$
- This distribution has no "memory" — the probability of succeeding on the *next* try is always $p$, no matter how many times you've already failed. (This is called the **memoryless property**, and it's a favorite trick question in interviews.)


```python

def geometric_explorer(p=0.3, max_k=20, num_simulated_trials=20):
    k_values = np.arange(1, max_k + 1)
    pmf = stats.geom.pmf(k_values, p)

    rng = np.random.default_rng(3)
    simulated_waits = rng.geometric(p, size=num_simulated_trials)

    fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

    axes[0].bar(k_values, pmf, color='#9467bd', edgecolor='white')
    axes[0].axvline(1/p, color='red', ls='--', label=f'Expected wait = 1/p = {1/p:.1f}')
    axes[0].set_xlabel('Trial number of first success (k)')
    axes[0].set_ylabel('P(X = k)')
    axes[0].set_title(f'Geometric Distribution (p={p})')
    axes[0].legend()

    axes[1].bar(range(1, num_simulated_trials + 1), simulated_waits, color='#8c564b')
    axes[1].axhline(simulated_waits.mean(), color='red', ls='--',
                     label=f'Avg wait observed = {simulated_waits.mean():.1f}')
    axes[1].set_xlabel('Simulated trainee #')
    axes[1].set_ylabel('Attempts until first success')
    axes[1].set_title(f'{num_simulated_trials} simulated trainees, each attempting until success')
    axes[1].legend()

    plt.tight_layout()
    plt.show()

    print(f"Theoretical expected number of attempts until first success = 1/p = {1/p:.2f}")
    print(f"Average across {num_simulated_trials} simulated trainees = {simulated_waits.mean():.2f}")

interact(
    geometric_explorer,
    p=widgets.FloatSlider(min=0.05, max=0.95, step=0.05, value=0.3, description='P(success):'),
    max_k=widgets.IntSlider(min=5, max=50, step=1, value=20, description='Show up to k:'),
    num_simulated_trials=widgets.IntSlider(min=5, max=200, step=5, value=20, description='# trainees:')
);
```

Hypothesis testing
### The core ideas

- **Null hypothesis $H_0$:** the boring, default explanation ("this coin is fair," "this drug does nothing," "these two methods perform the same").
- **Alternative hypothesis $H_1$:** what you suspect is actually true.
- **Significance level $\alpha$:** how much risk of a false conviction you're willing to accept — conventionally $\alpha = 0.05$ (5%).
- **p-value:** the probability of seeing evidence *at least this extreme*, **assuming $H_0$ is true**. Small p-value = the data would be a very unlikely coincidence under $H_0$ = reject $H_0$.
- **Decision rule:** if $p < \alpha$, reject $H_0$. Otherwise, "fail to reject" $H_0$ (never "accept" it — absence of evidence isn't evidence of absence).

### The two ways a verdict can go wrong

| | $H_0$ is actually True | $H_0$ is actually False |
|---|---|---|
| **You reject $H_0$** | ❌ Type I Error (false positive) — convicting an innocent man | ✅ Correct — caught the real effect |
| **You fail to reject $H_0$** | ✅ Correct — let an innocent man go | ❌ Type II Error (false negative) — a real effect slipped past you |

$P(\text{Type I Error}) = \alpha$. $P(\text{Type II Error}) = \beta$, and $1-\beta$ is called **statistical power** — your test's ability to catch a real effect when one exists.

Below, $H_0$'s distribution (grey) represents "no real effect." $H_1$'s distribution (orange) represents "there really is an effect of this size." Slide the decision threshold and watch the Type I / Type II error regions trade off against each other.


```python

def type1_type2_explorer(effect_size=1.5, sample_n=20, threshold=1.645):
    se = 1 / np.sqrt(sample_n) * np.sqrt(sample_n)  # standardized units; se scales the spread
    se = 1.0  # working in standardized (z) units for clarity
    x = np.linspace(-4, 4 + effect_size, 400)

    h0_pdf = stats.norm.pdf(x, 0, se)
    h1_pdf = stats.norm.pdf(x, effect_size, se)

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(x, h0_pdf, color='grey', lw=2, label='Distribution under $H_0$ (no effect)')
    ax.plot(x, h1_pdf, color='#ff7f0e', lw=2, label='Distribution under $H_1$ (true effect)')

    # Type I error: area under H0 beyond threshold
    x1 = np.linspace(threshold, 4 + effect_size, 200)
    ax.fill_between(x1, stats.norm.pdf(x1, 0, se), color='red', alpha=0.4, label='Type I error (α)')

    # Type II error: area under H1 below threshold
    x2 = np.linspace(-4, threshold, 200)
    ax.fill_between(x2, stats.norm.pdf(x2, effect_size, se), color='blue', alpha=0.3, label='Type II error (β)')

    ax.axvline(threshold, color='black', ls='--', label=f'Decision threshold = {threshold}')
    ax.set_title('Type I vs Type II Error Trade-off')
    ax.set_xlabel('Standardized test statistic')
    ax.legend(loc='upper left', fontsize=9)
    plt.tight_layout()
    plt.show()

    alpha = 1 - stats.norm.cdf(threshold, 0, se)
    beta = stats.norm.cdf(threshold, effect_size, se)
    power = 1 - beta
    print(f"α (Type I error rate)  = {alpha:.4f}")
    print(f"β (Type II error rate) = {beta:.4f}")
    print(f"Power (1 - β)          = {power:.4f}")

interact(
    type1_type2_explorer,
    effect_size=widgets.FloatSlider(min=0.0, max=4.0, step=0.1, value=1.5, description='True effect size:'),
    sample_n=widgets.fixed(2000),
    threshold=widgets.FloatSlider(min=-2.0, max=4.0, step=0.05, value=1.645, description='Decision threshold:')
);
```


- Move the `threshold` to the right — the red Type I region shrinks (fewer false convictions) but the blue Type II region grows (more real effects slip past you). You cannot shrink both at once by moving the threshold alone.
- Increase `effect_size` — with a bigger, more obvious real effect, the two curves separate and both error types shrink together. **Bigger effects are easier to detect.**
- **The only way to reduce both errors simultaneously is a bigger sample size** — larger $n$ narrows both curves without moving them, which is exactly why Case #11 onward will let you dial `n` directly.

## 📏 Case #11: The Baseline Check

The Academy's obstacle course has always taken about **300 seconds** to finish, on average. A new group of recruits just trained under a *revised* method. The question is simple: **did the new method actually change the time, or did this group just happen to be a little faster/slower by chance?**

### The core idea

We're comparing **one group's average** against an **already-known number** (300 seconds). This is called a **one-sample test**. (like how we compare the probabilties of the heights of people in different nation)

There are two versions, and which one you use depends on one thing: *do you know the true spread (standard deviation) of the whole population, or not?*

**Z-test** — use this if you already know the population's standard deviation, or your sample is large (30+ is a common rule of thumb):

$$z = \frac{\bar{x} - \mu_0}{\sigma / \sqrt{n}}$$

In plain words: *"how many standard errors away is my group's average from the expected 300 seconds?"*

**T-test** — use this instead when you *don't* know the population's true standard deviation (which is almost always the case in real life). You estimate the spread from your own sample, and because that estimate adds a bit of extra uncertainty, you use the **t-distribution** instead of the normal one — it's shaped like a bell curve but with slightly fatter tails, to be a little more cautious with small samples:

**One sample T-test** One group of measurements, compared to a known/assumed number → one-sample.
**Two sample T-test** - Two groups of measurements, compared to each other


$$t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}, \qquad df = n-1$$

**One thing to remember:** as your sample size grows, the t-distribution starts looking more and more like the normal distribution — the "extra caution" matters less once you have plenty of data. This is the same idea from the Central Limit Theorem (Case #6) showing up again.

```python

def one_sample_test_explorer(test_type='T-test (σ unknown)', mu0=300, true_mean=300, true_std=40, n=25, alpha=0.05):
    rng = np.random.default_rng(11)
    sample = rng.normal(true_mean, true_std, size=n)
    x_bar = sample.mean()
    s = sample.std(ddof=1)

    if test_type == 'Z-test (σ known)':
        se = true_std / np.sqrt(n)
        stat = (x_bar - mu0) / se
        p_value = 2 * (1 - stats.norm.cdf(abs(stat)))
        dist_label, crit = 'Normal (z)', stats.norm.ppf(1 - alpha/2)
        x = np.linspace(-4, 4, 400)
        curve = stats.norm.pdf(x)
    else:
        se = s / np.sqrt(n)
        stat = (x_bar - mu0) / se
        df = n - 1
        p_value = 2 * (1 - stats.t.cdf(abs(stat), df)) # 2 tailed hence multiply by 2
        dist_label, crit = f'Student t (df={df})', stats.t.ppf(1 - alpha/2, df)
        x = np.linspace(-4, 4, 400)
        curve = stats.t.pdf(x, df)

    fig, ax = plt.subplots(figsize=(9, 4.5))
    ax.plot(x, curve, color='#1f77b4', lw=2, label=f'{dist_label} distribution under $H_0$')
    reject_region_hi = x[x >= crit]
    reject_region_lo = x[x <= -crit]
    ax.fill_between(reject_region_hi, curve[x >= crit], color='red', alpha=0.3, label=f'Rejection region (α={alpha})')
    ax.fill_between(reject_region_lo, curve[x <= -crit], color='red', alpha=0.3)
    ax.axvline(stat, color='black', ls='--', lw=2, label=f'Observed statistic = {stat:.2f}')
    ax.set_title(f'One-Sample {test_type.split()[0]}-Test:  $H_0$: μ = {mu0}')
    ax.legend(fontsize=9)
    plt.tight_layout()
    plt.show()

    decision = "REJECT H0" if p_value < alpha else "FAIL TO REJECT H0"
    print(f"Sample mean (x̄) = {x_bar:.2f}   Sample size (n) = {n}")
    print(f"Test statistic   = {stat:.4f}")
    print(f"p-value          = {p_value:.4f}")
    print(f"Decision at α={alpha}: {decision}")

interact(
    one_sample_test_explorer,
    test_type=widgets.Dropdown(options=['Z-test (σ known)', 'T-test (σ unknown)'], description='Test type:'),
    mu0=widgets.FloatSlider(min=250, max=350, step=1, value=300, description='H0: μ0 ='),
    true_mean=widgets.FloatSlider(min=250, max=350, step=1, value=310, description='True mean (secret):'),
    true_std=widgets.FloatSlider(min=10, max=80, step=5, value=40, description='True σ (secret):'),
    n=widgets.IntSlider(min=3, max=200, step=1, value=25, description='Sample size (n):'),
    alpha=widgets.SelectionSlider(options=[0.01, 0.05, 0.10], value=0.05, description='α:')
);
```

**Notes**

- Set `true_mean = mu0 = 300` (no real difference) and re-run several times — you'll see p-values bounce around, occasionally dropping below 0.05 purely by chance. That's exactly your `α` — a 5% false-alarm rate is *built into* the method, not a flaw.
- Now separate them (`true_mean = 320`, `mu0 = 300`) and watch how increasing `n` makes the test far more likely to correctly reject $H_0$ — more data means more statistical power (straight from Case #10).
- **Exercise:** With a small `n` (say 5) and a big gap between `true_mean` and `mu0`, compare the t-test's rejection region width to the z-test's — the t-distribution's fatter tails demand stronger evidence when your sample is small.


# P-Value

If nothing interesting were actually happening (H₀ is true), how surprising is the data I actually got?"


**Small p-value (like 0.01) **→ "if there were really no effect, seeing data this extreme would be very rare/surprising — so maybe there is something real going on."
**Large p-value (like 0.60) **→ "if there were really no effect, data like this would be totally normal and unsurprising — no strong reason to doubt the boring explanation."

It is not "the probability that H₀ is true" — that's a very common misreading. It's "the probability of seeing evidence this extreme (or more), assuming H₀ is true."

##  Which Test Should I Use?

### Step 1 — What kind of outcome are you measuring?

- **Categorical** (pass/fail, chosen track, yes/no) → go to **Categorical outcomes**
- **Numeric** (score, time, height) → go to **Numeric outcomes**

---

### Categorical outcomes

- **Comparing observed counts against an expected split?**
  → **Chi-square Goodness-of-Fit** *(Case 13)*
- **Checking if two categorical variables are related?**
  → **Chi-square Test of Independence** *(Case 13)*

---

### Numeric outcomes

**How many groups are you comparing?**

- **1 group vs. a known/fixed value**
  - Population std dev (σ) known, or large sample? → **Z-test**
  - σ unknown (estimate from sample)? → **One-Sample T-test** *(Case 11)*

- **2 groups**
  - Same subjects measured twice (before/after)? → **Paired T-test** *(Case 12)*
    - Data not normal / small sample? → use **Wilcoxon Signed-Rank Test** instead
  - Two different, unrelated groups? → **Independent T-test (Welch's)** *(Case 12)*
    - Data not normal / small sample? → use **Mann-Whitney U Test** instead
    - Similar variances → Student's t-test (assumes equal variance)
    - Different variances → Welch's t-test
- **3+ groups**
  - One factor only (e.g. just "method")? → **One-Way ANOVA** *(Case 14)*
    - Data not normal / small sample? → use **Kruskal-Wallis Test** instead
    - ANOVA significant? → run **Tukey HSD** post-hoc to find which pairs differ
  - Two factors together (e.g. "method" AND "experience")? → **Two-Way ANOVA** *(Case 15)*

- **Relationship between two numeric variables** (not "different groups" — e.g. does study time relate to score?)
  - Roughly linear relationship? → **Pearson Correlation**
  - Non-linear but still consistently increasing/decreasing? → **Spearman Correlation**

---

### 🔍 Before trusting any T-test / ANOVA result, quickly check:

| Assumption | Check with | If it fails... |
|---|---|---|
| Data is roughly normal | `stats.shapiro(x)` | Switch to the non-parametric alternative above |
| Groups have similar variance | `stats.levene(x1, x2)` | Use Welch's version (unequal variance) instead of Student's |

---

**One-liner to remember:**
> *Count or measure → how many groups → known or estimated spread → normal or not.*
> *Answer those, and the test picks itself.*


## Case #12: The Two Academies

Two training methods are being piloted — **Method A** (traditional) and **Method B** (revised). The obvious question: *does one produce better scores than the other?*

There are two distinct ways to design this comparison, and they call for two different tests.

### Independent two-sample t-test

Different trainees in each group (unrelated to each other):

$$t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\dfrac{s_1^2}{n_1} + \dfrac{s_2^2}{n_2}}}$$

(This is **Welch's t-test** — it doesn't assume the two groups have equal variance, which is the safer default in practice.)

### Paired t-test

The *same* trainees, measured twice (e.g. before vs. after a program) — each person acts as their own control, which removes person-to-person variability and makes the test more powerful:

$$t = \frac{\bar{d}}{s_d / \sqrt{n}}, \qquad d_i = x_{i,\text{after}} - x_{i,\text{before}}$$

Choosing the right one matters — using an independent test on paired data throws away useful information, and vice versa produces nonsense.


```python

def two_sample_test_explorer(design='Independent (two different groups)',
                              mean_A=70, mean_B=75, std_A=10, std_B=10, n=30, alpha=0.05):
    rng = np.random.default_rng(12)

    if design == 'Independent (two different groups)':
        group_A = rng.normal(mean_A, std_A, n)
        group_B = rng.normal(mean_B, std_B, n)
        stat, p_value = stats.ttest_ind(group_A, group_B, equal_var=False)
        test_name = "Welch's Independent Two-Sample T-test"
    else:
        # paired: simulate "before" scores, then an individual improvement + noise
        before = rng.normal(mean_A, std_A, n)
        improvement = (mean_B - mean_A)
        after = before + improvement + rng.normal(0, std_B * 0.4, n)
        group_A, group_B = before, after
        stat, p_value = stats.ttest_rel(group_B, group_A)
        test_name = "Paired T-test"

    fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))
    axes[0].hist(group_A, bins=15, alpha=0.6, label='Group A / Before', color='#1f77b4')
    axes[0].hist(group_B, bins=15, alpha=0.6, label='Group B / After', color='#ff7f0e')
    axes[0].axvline(group_A.mean(), color='#1f77b4', ls='--')
    axes[0].axvline(group_B.mean(), color='#ff7f0e', ls='--')
    axes[0].set_title('Score Distributions')
    axes[0].legend()

    axes[1].boxplot([group_A, group_B], labels=['A / Before', 'B / After'])
    axes[1].set_title(f'{test_name}\nt = {stat:.3f}, p = {p_value:.4f}')

    plt.tight_layout()
    plt.show()

    decision = "REJECT H0 (means differ significantly)" if p_value < alpha else "FAIL TO REJECT H0 (no significant difference detected)"
    print(f"Mean A/Before = {group_A.mean():.2f}   Mean B/After = {group_B.mean():.2f}")
    print(f"t-statistic = {stat:.4f}   p-value = {p_value:.4f}")
    print(f"Decision at α={alpha}: {decision}")

interact(
    two_sample_test_explorer,
    design=widgets.Dropdown(options=['Independent (two different groups)', 'Paired (same trainees, before/after)'], description='Design:'),
    mean_A=widgets.FloatSlider(min=50, max=90, step=1, value=70, description='Mean A:'),
    mean_B=widgets.FloatSlider(min=50, max=90, step=1, value=75, description='Mean B:'),
    std_A=widgets.FloatSlider(min=2, max=25, step=1, value=10, description='Std A:'),
    std_B=widgets.FloatSlider(min=2, max=25, step=1, value=10, description='Std B:'),
    n=widgets.IntSlider(min=5, max=200, step=5, value=30, description='n per group:'),
    alpha=widgets.SelectionSlider(options=[0.01, 0.05, 0.10], value=0.05, description='α:')
);
```


** Notes**

- With `mean_A = mean_B` and `design = Independent`, run it a few times — most trials should fail to reject $H_0$, exactly as expected when there's genuinely no difference.
- Switch to `Paired` with the same mean gap — notice the p-value tends to be *smaller* (more significant) than the independent design with identical numbers. Pairing removes individual-to-individual noise, boosting power.
- **Exercise:** Set a small mean gap (`mean_A=70, mean_B=73`) with high `std` and small `n` — the test likely fails to detect a real (small) difference. Now raise `n` to 150 — same true gap, but now it's detected. This is the sample-size/power lesson from Case #10, live again.


## ⚖️ The 3-Step Rule for Accepting or Rejecting H₀

### Step 1 — Pick your significance level (α), *before* looking at any results

Usually **α = 0.05** (5%). This is your "how surprising does the data need to be before I stop believing H₀?" threshold — decided in advance, never after seeing the p-value.

### Step 2 — Run the test and get your p-value

The p-value is the probability of seeing data this extreme (or more), **assuming H₀ is true**.

### Step 3 — Compare the p-value to α

| Condition | Decision |
|---|---|
| $p < \alpha$ | **REJECT $H_0$** (result is statistically significant) |
| $p \ge \alpha$ | **FAIL TO REJECT $H_0$** (not enough evidence) |
p-value = (1-cdf)
---

### 🧠 Memory hook

> **"Small p, reject H₀. Big p, stick with H₀."**

---

### ⚠️ Two things to keep in mind

- **"Fail to reject" is not "accept."** A big p-value doesn't prove $H_0$ is true — it just means the data didn't give enough evidence to knock it down. (Same logic as a "not guilty" verdict — it doesn't mean "proven innocent.")
- **α must be chosen *before* seeing the data.** Picking α = 0.05 and then switching to α = 0.10 because your p-value came out to 0.07 defeats the entire purpose of the test.


## Case #13: The Category Files

Not all data is numeric — sometimes you're working with **categories**. The Chi-Square ($\chi^2$) family handles exactly this.

### Two flavors

**1. Goodness-of-fit test** — does an observed distribution of categories match an *expected* distribution?
> *"New recruits pick one of 4 specialty tracks. Is the split actually uniform (25% each), or do some tracks get picked more?"*

**2. Test of independence** — are two categorical variables related, or independent?
> *"Is passing the final exam related to which training method a recruit used?"*

Both use the same statistic, comparing observed counts $O_i$ to expected counts $E_i$:

$$\chi^2 = \sum_i \frac{(O_i - E_i)^2}{E_i}$$

A large $\chi^2$ means the observed counts deviate a lot from what you'd expect under $H_0$ ("no preference" / "no relationship") — which translates into a small p-value.


```python

def goodness_of_fit_explorer(count_A=30, count_B=25, count_C=20, count_D=25, alpha=0.05):
    observed = np.array([count_A, count_B, count_C, count_D])
    labels = ['Track A', 'Track B', 'Track C', 'Track D']
    expected = np.full(4, observed.sum() / 4)  # H0: uniform split

    chi2_stat, p_value = stats.chisquare(observed, f_exp=expected)

    fig, ax = plt.subplots(figsize=(8, 4.5))
    x = np.arange(4)
    width = 0.35
    ax.bar(x - width/2, observed, width, label='Observed', color='#1f77b4')
    ax.bar(x + width/2, expected, width, label='Expected (uniform)', color='#7f7f7f')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.set_ylabel('Number of recruits')
    ax.set_title(f'Goodness-of-Fit Test\nχ² = {chi2_stat:.3f}, p = {p_value:.4f}')
    ax.legend()
    plt.tight_layout()
    plt.show()

    decision = "REJECT H0 (specialty choice is NOT uniform)" if p_value < alpha else "FAIL TO REJECT H0 (consistent with uniform choice)"
    print(f"Decision at α={alpha}: {decision}")

interact(
    goodness_of_fit_explorer,
    count_A=widgets.IntSlider(min=0, max=100, step=1, value=30, description='Track A:'),
    count_B=widgets.IntSlider(min=0, max=100, step=1, value=25, description='Track B:'),
    count_C=widgets.IntSlider(min=0, max=100, step=1, value=20, description='Track C:'),
    count_D=widgets.IntSlider(min=0, max=100, step=1, value=25, description='Track D:'),
    alpha=widgets.SelectionSlider(options=[0.01, 0.05, 0.10], value=0.05, description='α:')
);
```

```python

def independence_test_explorer(passA=40, failA=10, passB=30, failB=20, alpha=0.05):
    table = np.array([[passA, failA], [passB, failB]])
    df_table = pd.DataFrame(table, index=['Method A', 'Method B'], columns=['Pass', 'Fail'])
    display(df_table)

    chi2_stat, p_value, dof, expected = stats.chi2_contingency(table)

    fig, ax = plt.subplots(figsize=(7, 4.5))
    im = ax.imshow(table, cmap='Blues')
    for i in range(2):
        for j in range(2):
            ax.text(j, i, f"{table[i,j]}\n(exp {expected[i,j]:.1f})", ha='center', va='center', fontsize=11)
    ax.set_xticks([0, 1]); ax.set_xticklabels(['Pass', 'Fail'])
    ax.set_yticks([0, 1]); ax.set_yticklabels(['Method A', 'Method B'])
    ax.set_title(f'Test of Independence\nχ² = {chi2_stat:.3f}, df={dof}, p = {p_value:.4f}')
    plt.tight_layout()
    plt.show()

    decision = "REJECT H0 (pass rate DEPENDS on method)" if p_value < alpha else "FAIL TO REJECT H0 (pass rate independent of method)"
    print(f"Decision at α={alpha}: {decision}")

interact(
    independence_test_explorer,
    passA=widgets.IntSlider(min=0, max=100, step=1, value=40, description='A: Pass'),
    failA=widgets.IntSlider(min=0, max=100, step=1, value=10, description='A: Fail'),
    passB=widgets.IntSlider(min=0, max=100, step=1, value=30, description='B: Pass'),
    failB=widgets.IntSlider(min=0, max=100, step=1, value=20, description='B: Fail'),
    alpha=widgets.SelectionSlider(options=[0.01, 0.05, 0.10], value=0.05, description='α:')
);
```


- In the goodness-of-fit widget, set all four counts equal (e.g. 25 each) — p-value should be large (fails to reject uniformity), exactly as expected.
- In the independence widget, set the pass/fail ratio *identical* for both methods (e.g. `40/10` and `40/10`) — large p-value, no detected relationship. Now skew Method B's fail rate way up — watch χ² grow and p shrink.
- **Rule of thumb:** Chi-square tests need reasonably large expected counts in every cell (typically ≥5) to be reliable — with tiny samples, look up **Fisher's exact test** instead.


## 🏆 Case #14: The Method Showdown

Case #12 compared **two** training methods. What if the Academy is piloting **three or more**? You could run pairwise t-tests between every pair, but that inflates your false-positive rate (more comparisons = more chances for a lucky fluke). The right tool is **ANOVA** — Analysis of Variance.

### The core idea

One-way ANOVA tests whether **at least one** group mean differs from the others, by comparing *between-group variance* to *within-group variance*:

$$F = \frac{\text{Mean Square Between groups (MSB)}}{\text{Mean Square Within groups (MSW)}}$$

- $H_0$: all group means are equal ($\mu_1 = \mu_2 = \mu_3 = \dots$)
- $H_1$: at least one group mean differs
- A large $F$ means groups differ from each other *more* than individuals differ within their own group — a signal that something real is going on.

ANOVA only tells you *that* a difference exists somewhere, not *which* groups differ — for that, a **post-hoc test** like Tukey's HSD is used, which we run automatically here whenever the ANOVA is significant.


```python

def one_way_anova_explorer(mean_1=70, mean_2=73, mean_3=78, std=8, n_per_group=25, alpha=0.05):
    rng = np.random.default_rng(14)
    g1 = rng.normal(mean_1, std, n_per_group)
    g2 = rng.normal(mean_2, std, n_per_group)
    g3 = rng.normal(mean_3, std, n_per_group)

    f_stat, p_value = stats.f_oneway(g1, g2, g3)

    fig, ax = plt.subplots(figsize=(8, 4.5))
    ax.boxplot([g1, g2, g3], labels=['Method 1', 'Method 2', 'Method 3'])
    ax.set_title(f'One-Way ANOVA\nF = {f_stat:.3f}, p = {p_value:.4f}')
    ax.set_ylabel('Score')
    plt.tight_layout()
    plt.show()

    decision = "REJECT H0 (at least one method differs)" if p_value < alpha else "FAIL TO REJECT H0 (no significant difference detected)"
    print(f"Group means: {g1.mean():.2f}, {g2.mean():.2f}, {g3.mean():.2f}")
    print(f"F-statistic = {f_stat:.4f}   p-value = {p_value:.4f}")
    print(f"Decision at α={alpha}: {decision}")

    if p_value < alpha:
        res = stats.tukey_hsd(g1, g2, g3)
        print("\nPost-hoc Tukey HSD (which pairs actually differ?):")
        print(res)

interact(
    one_way_anova_explorer,
    mean_1=widgets.FloatSlider(min=50, max=90, step=1, value=70, description='Mean 1:'),
    mean_2=widgets.FloatSlider(min=50, max=90, step=1, value=73, description='Mean 2:'),
    mean_3=widgets.FloatSlider(min=50, max=90, step=1, value=78, description='Mean 3:'),
    std=widgets.FloatSlider(min=2, max=25, step=1, value=8, description='Std dev:'),
    n_per_group=widgets.IntSlider(min=5, max=100, step=5, value=25, description='n per group:'),
    alpha=widgets.SelectionSlider(options=[0.01, 0.05, 0.10], value=0.05, description='α:')
);
```

- Set all three means equal (e.g. 70, 70, 70) — the F-statistic should hover near 1 and p-value should be large.
- Now spread them out (60, 75, 90) — F climbs, p shrinks, and the Tukey HSD output tells you exactly which pairs differ significantly (not just "something is different somewhere").
- **Exercise:** Set means to 70, 72, 90 — ANOVA should reject $H_0$, but Tukey's output will likely show Method 3 differs from both 1 and 2, while 1 vs 2 remains inconclusive. This is exactly the extra granularity ANOVA alone can't give you.


## Case #15: The Two-Factor Mystery

Final advanced case. The Academy suspects that scores depend not just on **training method**, but also on **experience level** (Junior vs Senior) — and possibly on the *combination* of the two (maybe Method B works great for Seniors but does nothing for Juniors).

### The core idea

**Two-way ANOVA** tests three things simultaneously:
- **Main effect of Factor 1** (Method): does method matter, ignoring experience?
- **Main effect of Factor 2** (Experience): does experience matter, ignoring method?
- **Interaction effect**: does the *effect of method depend on experience level*? (This is often the most interesting finding — and the one a one-way ANOVA could never catch.)

We'll use `statsmodels` to fit this properly, the same way it's done in real research.


```python

import statsmodels.api as sm
from statsmodels.formula.api import ols

def two_way_anova_explorer(method_effect=5.0, experience_effect=8.0, interaction_effect=0.0, std=8, n_per_cell=20, alpha=0.05):
    rng = np.random.default_rng(15)
    rows = []
    base = 70
    for method in ['A', 'B']:
        for exp in ['Junior', 'Senior']:
            m_eff = method_effect if method == 'B' else 0
            e_eff = experience_effect if exp == 'Senior' else 0
            inter = interaction_effect if (method == 'B' and exp == 'Senior') else 0
            scores = rng.normal(base + m_eff + e_eff + inter, std, n_per_cell)
            for s in scores:
                rows.append({'Method': method, 'Experience': exp, 'Score': s})

    df = pd.DataFrame(rows)

    model = ols('Score ~ C(Method) * C(Experience)', data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    display(anova_table.round(4))

    # Interaction plot
    means = df.groupby(['Method', 'Experience'])['Score'].mean().unstack()
    fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

    means.plot(marker='o', ax=axes[0])
    axes[0].set_title('Interaction Plot (parallel lines = no interaction)')
    axes[0].set_ylabel('Mean Score')

    df.boxplot(column='Score', by=['Method', 'Experience'], ax=axes[1])
    axes[1].set_title('Score by Method × Experience')
    plt.suptitle('')
    plt.tight_layout()
    plt.show()

    print(f"\nDecisions at α={alpha}:")
    for factor in ['C(Method)', 'C(Experience)', 'C(Method):C(Experience)']:
        p = anova_table.loc[factor, 'PR(>F)']
        verdict = "SIGNIFICANT" if p < alpha else "not significant"
        print(f"  {factor:30s}  p = {p:.4f}  ->  {verdict}")

interact(
    two_way_anova_explorer,
    method_effect=widgets.FloatSlider(min=-20, max=20, step=1, value=5, description='Method effect:'),
    experience_effect=widgets.FloatSlider(min=-20, max=20, step=1, value=8, description='Experience effect:'),
    interaction_effect=widgets.FloatSlider(min=-20, max=20, step=1, value=0, description='Interaction effect:'),
    std=widgets.FloatSlider(min=2, max=25, step=1, value=8, description='Std dev:'),
    n_per_cell=widgets.IntSlider(min=5, max=100, step=5, value=20, description='n per cell:'),
    alpha=widgets.SelectionSlider(options=[0.01, 0.05, 0.10], value=0.05, description='α:')
);
```

| Test | Use it when... | scipy / statsmodels call |
|---|---|---|
| **Shapiro-Wilk** | Checking if a sample looks normally distributed (many tests above assume this) | `stats.shapiro(x)` |
| **Levene's test** | Checking if two/more groups have equal variance (an assumption of the classic t-test/ANOVA) | `stats.levene(x1, x2)` |
| **Mann-Whitney U** | Non-parametric alternative to the independent t-test (data isn't normal, or is ordinal) | `stats.mannwhitneyu(x1, x2)` |
| **Wilcoxon signed-rank** | Non-parametric alternative to the paired t-test | `stats.wilcoxon(x1, x2)` |
| **Kruskal-Wallis** | Non-parametric alternative to one-way ANOVA | `stats.kruskal(x1, x2, x3)` |
| **Fisher's exact test** | Like chi-square independence, but reliable with very small sample counts | `stats.fisher_exact(table)` |
| **Pearson correlation** | Strength of *linear* relationship between two continuous variables | `stats.pearsonr(x, y)` |
| **Spearman correlation** | Strength of *monotonic* (not necessarily linear) relationship; robust to outliers | `stats.spearmanr(x, y)` |
| **F-test for equal variances** | Directly comparing two variances | `stats.levene` or `stats.bartlett` |
